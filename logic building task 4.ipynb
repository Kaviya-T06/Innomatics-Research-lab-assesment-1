```python
def recharge_plan(amount, valid_plan):
    if amount < 50:
        return "Amount is not sufficient"
    allowed_plans = [199, 299, 399, 599]
    while valid_plan not in allowed_plans:
        print("Plan is not valid")
        valid_plan = int(input("Enter valid plan (199/299/399/599): "))
    return "Recharge successful with plan " + str(valid_plan)
amount = int(input("Enter amount: "))
valid_plan = int(input("Enter valid plan: "))
res = recharge_plan(amount, valid_plan)
print(res)
```

```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert ⚠️")
        else:
            print(f"{product}: Stock OK ✅")
inventory = {
    "Rice": 20,
    "Wheat": 10,
    "Sugar": 5,
    "Oil": 25,
    "Salt": 15
}

check_inventory(inventory)
```

```python
def marks_cal(stud_marks):
    count=0
    for i in stud_marks:
        count+=i
    avg=count/len(stud_marks)
    print("count:",count)
    print("Average:",avg)
    if avg>=50:
        print("Pass")
    else:
        print("fail")
stud_marks=[70,80,90,65,89]
res=marks_cal(stud_marks)
print(res)
```

```python
def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km_rate = 12
    fare = base_fare + (distance * per_km_rate)
    if is_peak:
        fare += fare * 0.25   # 25% extra
    return fare
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()
        if peak_input == "yes":
            is_peak = True
        elif peak_input == "no":
            is_peak = False
        else:
            print("Invalid input for peak hour. Try again.")
            continue
        total_fare = calculate_fare(distance, is_peak)
        print(f"Total Cab Fare: ₹{total_fare}")
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using Cab Fare Calculator!")
            break
    except ValueError:
        print("Invalid input! Please enter numeric value for distance.")
```

```python
def att_rate(att):
    total_days = len(att)
    present_count = 0
    for i in att:
        if i == "P":
            present_count += 1
    percentage = (present_count / total_days) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
att = ["P", "A", "P", "P", "A"]
result = att_rate(att)
print("Attendance Percentage:", (att.count("P") / len(att)) * 100)
print("Result:", result)

    
```

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"  
    has_digit = False
    has_special = False
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"
password = input("Enter your password: ")
result = check_password_strength(password)
print(result)

    
```
