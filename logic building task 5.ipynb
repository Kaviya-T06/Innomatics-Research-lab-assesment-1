```python
def analyze_engagement(likes_list):
    total_likes = 0

   
    for likes in likes_list:
        total_likes += likes

    print("Total Likes:", total_likes)

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")
likes_list = [400, 350, 300]

analyze_engagement(likes_list)
```

```python
def rainfall_checker(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain
    average = total / len(rainfall_list)
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)
rainfall_data = [70, 75, 80, 65, 70]
rainfall_checker(rainfall_data, 70)
```

```python
def check_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
users = ["kaviya", "rahul", "kaviya", "anil"]
check_duplicates(users)
```

```python
def appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    print("Patient Age:", age)
    print("Eligibility Status:", status)
appointment_eligibility(21)
```

```python
def premium_crops(price_list):
    premium = []
    for price in price_list:
        if price > 2000:
            premium.append(price)
    print("Premium Crops:", premium)
prices = [1500, 2500, 3200, 1800]
premium_crops(prices)
```

```python
def system_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print("Error Count:", error_count)
    print("System Status:", status)
system_health(7)
```

```python
def transaction_checker(amount):
    daily_limit = 50000
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)
transaction_checker(60000)
```

```python
def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = 0
    for status in attendance_list:
        if status == 1:   # 1 means present
            attended += 1
    percentage = (attended / total_classes) * 100
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)
attendance = [1,1,1,0,1,1,1,1,0,1]
attendance_checker(attendance)
```

```python
def electricity_bill(units):
    bill = 0 
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    return bill, usage
units_consumed = 250
bill_amount, usage_status = electricity_bill(units_consumed)
print("Total Bill:", bill_amount)
print("Usage Level:", usage_status)
```
