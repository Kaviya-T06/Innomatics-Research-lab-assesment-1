```python
employees = {"Ravi": 92, "Anita": 88, "Kiran": 92, "Suresh": 85}#program 1
highest = max(employees.values())
performers = []
for name, score in employees.items():
    if score == highest:
        performers.append(name)
print("Top performers eligible for bonus:", ", ".join(performers), f"(Score: {highest})")
```

```python
a = "Buy mobile phone buy phone calling"#program 2
a = a.lower()
words = a.split()
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1
result = {}
for word, count in freq.items():
    if count > 1:
        result[word] = count
print(result)
```

```python
emails=["ravi@gmail.com:","anita@yahoo.com:","Kirn@gmail.com","Suresh@gmail.com","meena@yahoo.com"]#program 4
count=0
total=0
for char in emails:
    if "@gmail.com" in char :
        count+=1
    elif "@yahoo.com" in char:
        total+=1
percentage=int(count/len(emails)*100)
percent=int(total/len(emails)*100)
print("gmail.com:",(percentage),"%")
print("yahoo.com:",(percent),"%")
        
        
```

```python
sales=[1200,1500,900,2200,1400,3000]#program 5
average=sum(sales)/len(sales)
for index, value in enumerate(sales, start=1):
    if value > average:
        print(f"Day {index}: {value}")
```
