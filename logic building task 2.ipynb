```python
employees={"Ravi":75000,"Anita":68000,"Kiran":72000}#2nd problem
highest=employees["Ravi"]
for i in employees:
    if employees[i]>highest:
        highest=employees[i]
print("highest Salary:",highest)
```

```python
numbers=[45,22,89,10,66]#3rd problem
maximum=numbers[0]
minimum=numbers[0]
for i in numbers:
    if i>maximum:
        maximum=i
    if i<minimum:
        minimum=i
print("Max:",maximum)
print("Min:",minimum)
    
```

```python
text = "python"#7th problem
freq = {}
for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1
print(freq)
```
